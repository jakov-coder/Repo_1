```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_encoded,
                                                    y_encoded,
                                                    test_size = 0.20,
                                                    random_state = 5)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
pd.DataFrame(x_test, columns=x_encoded.columns)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0, penalty="elasticnet",
                                solver="saga", l1_ratio=0.9)
classifier.fit(x_train, y_train)
```

```python
pd.Series(np.squeeze(classifier.coef_), index=x_encoded.columns)
```

```python
pd.Series(np.squeeze(classifier.coef_), name="coefs", index=x_encoded.columns).sort_values().plot.barh()
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
y_pred = pd.Series(y_pred, name='Pred', index=y_test.index)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
results = pd.concat([y_test, y_pred], axis=1)
results
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
accuracy_score(y_test, y_pred)*100
```
